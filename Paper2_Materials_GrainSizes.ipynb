```python
#### Sand ####
#Unaveraged Sand

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

sand_df = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes\Paper2_sand.csv', header=1))
sand_gvals = sand_df[:,::2]
sand_vvals = sand_df[:,1::2]

plt.plot(sand_gvals, sand_vvals)
plt.title('Unaveraged Sand')
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()

#Averaged Sand

sand_df_avg = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes\Paper2_sand_avg.csv', header=1))
sand_gvals_avg = sand_df_avg[:,0]
sand_vvals_avg = sand_df_avg[:,1]
plt.plot(sand_gvals_avg, sand_vvals_avg)
plt.title('Averaged Sand')
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()
```

```python
#### Glass Beads ####

#Unaveraged Glass Beads
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

GB_df = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes\Paper2_GB.csv', header=1))
GB_gvals = GB_df[:,::2]
GB_vvals = GB_df[:,1::2]

plt.plot(GB_gvals, GB_vvals)
plt.title('Unaveraged Glass Beads')
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()

#Averaged Sand

GB_df_avg = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes\Paper2_GB_avg.csv', header=1))
GB_gvals_avg = GB_df_avg[:,0]
GB_vvals_avg = GB_df_avg[:,1]
plt.plot(GB_gvals_avg, GB_vvals_avg)
plt.title('Averaged Sand')
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()
```

```python
#### LHS-1 no fines ####
#Unaveraged

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

LHS1_df = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes\Paper2_LHS1NF.csv', header=1))
LHS1_gvals = LHS1_df[:,::2]
LHS1_vvals = LHS1_df[:,1::2]

plt.plot(LHS1_gvals, LHS1_vvals)
plt.title('Unaveraged LHS-1, no fines')
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()

#Averaged Sand

LHS1_df_avg = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes\Paper2_LHS1NF_avg.csv', header=1))
LHS1_gvals_avg = LHS1_df_avg[:,0]
LHS1_vvals_avg = LHS1_df_avg[:,1]
plt.plot(LHS1_gvals_avg, LHS1_vvals_avg)
plt.title('Averaged LHS-1, no fines')
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()
```

```python
#### LHS-1 w fines ####
#Unaveraged

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

LHS1_df_fines = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes\Paper2_LHS1_fines.csv', header=1))
LHS1_gvals_fines = LHS1_df_fines[:,::2]
LHS1_vvals_fines = LHS1_df_fines[:,1::2]

plt.plot(LHS1_gvals_fines, LHS1_vvals_fines)
plt.title('Unaveraged LHS-1, no fines')
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()

#Averaged LHS-1 with fines

LHS1_df_fines_avg = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes\Paper2_LHS1_fines_avg.csv', header=1))
LHS1_gvals_fines_avg = LHS1_df_fines_avg[:,0]
LHS1_vvals_fines_avg = LHS1_df_fines_avg[:,1]
plt.plot(LHS1_gvals_fines_avg, LHS1_vvals_fines_avg)
plt.title('Averaged LHS-1, no fines')
plt.xscale('log')
plt.xlabel('Grain Size(μm)')
plt.ylabel('Diff. % Volume(%)')
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

##### UFR
UFR_df = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes/UFR_avg.csv', header=1))
UFR_gvals = UFR_df[:,0]
UFR_vvals = UFR_df[:,1]

##### DFR
DFR_df = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes/DFR_avg.csv', header=1))
DFR_gvals = DFR_df[:,0]
DFR_vvals = DFR_df[:,1]

##### LPR
LPR_df = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes/LPR_avg.csv', header=1))
LPR_gvals = LPR_df[:,0]
LPR_vvals = LPR_df[:,1]

##### MR
MR_df = np.asarray(pd.read_csv('Paper2_Materials_GrainSizes/MR_avg.csv', header=1))
MR_gvals = MR_df[:,0]
MR_vvals = MR_df[:,1]

####plotting
plt.plot(UFR_gvals, UFR_vvals, label = 'unyielding frozen regolith', color = 'navy', linewidth=3)
plt.plot(DFR_gvals, DFR_vvals, label = 'ductile frozen regolith', color = 'darkgray', linewidth=3)
plt.plot(LPR_gvals, LPR_vvals, label = 'loose particle regolith', color = 'tan', linewidth=3)
plt.plot(MR_gvals, MR_vvals, label = 'moist regolith', color = 'blue', linewidth=3)


#### plot params
plt.xscale('log')
plt.xlabel('grain size (μm)', fontsize = 'xx-large')
plt.ylabel('volume (%)', fontsize = 'xx-large')
plt.legend(loc = 'upper left', fontsize = 'x-large', frameon = False)

plt.xticks(fontsize=14)  # Change x-axis text size
plt.yticks(fontsize=14)  # Change y-axis text size

plt.show()
```
